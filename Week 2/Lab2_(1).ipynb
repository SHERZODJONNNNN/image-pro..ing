```python
# Task 2

import cv2
import matplotlib.pyplot as plt

# Read image from Drive
img = cv2.imread("/content/drive/MyDrive/opencv/fruit_basket.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Add different intensity values
bright_30  = cv2.add(img_rgb, 30)
bright_100 = cv2.add(img_rgb, 100)

# Show results
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bright_30)
plt.title("Added +30")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(bright_100)
plt.title("Added +100")
plt.axis("off")

plt.show()
```

```python
# TAsk 3

import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("/content/drive/MyDrive/opencv/fruit_basket.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Task 3: Multiplication experiments
# Factor < 1 (Decreases contrast/darkens)
multiplied_05 = cv2.multiply(img_rgb, 0.5)

# Factor > 1 (Increases contrast/brightens)
multiplied_20 = cv2.multiply(img_rgb, 2.0)

# Show results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(multiplied_05)
plt.title("Multiplied by 0.5")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(multiplied_20)
plt.title("Multiplied by 2.0")
plt.axis("off")

plt.show()
```

```python
# Task 4

# Experimenting with factors 3.0 and 0.5
div_30 = cv2.divide(img_rgb, 3.0) # Factor > 1
div_05 = cv2.divide(img_rgb, 0.5) # Factor < 1

# Display Task 4
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
plt.subplot(1, 3, 2); plt.imshow(div_30); plt.title("Div 3.0 (Darker)"); plt.axis("off")
plt.subplot(1, 3, 3); plt.imshow(div_05); plt.title("Div 0.5 (Brighter)"); plt.axis("off")
plt.show()
```
