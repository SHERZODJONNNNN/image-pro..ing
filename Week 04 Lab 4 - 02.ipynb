```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = '/content/drive/MyDrive/opencv/download (1).jpeg'

def color_to_gray(image_path):
    img_color = cv2.imread(image_path)
    img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title('Input')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Output')
    plt.axis('off')
    plt.show()

def mirror_effect(image_path):
    img = cv2.imread(image_path)
    img_flipped = cv2.flip(img, 1)
    mirror = np.concatenate([img, img_flipped], axis=1)
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Input')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(mirror, cv2.COLOR_BGR2RGB))
    plt.title('Mirror')
    plt.axis('off')
    plt.show()

color_to_gray(image_path)
mirror_effect(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = '/content/drive/MyDrive/opencv/download (1).jpeg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Task 1 - Negative
negative = 255 - img
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(negative, cmap='gray'); plt.title('Negative'); plt.axis('off')
plt.show()

# Task 2a - Gamma curves non-normalized
gamma_values = [0.02, 0.2, 0.5, 0.8, 1, 2, 3, 3.5, 4]
x = np.arange(0, 256)
for g in gamma_values:
    plt.plot(x, x ** g, label=f'={g}')
plt.title('Non-Normalized (0-255)')
plt.legend(fontsize=7)
plt.show()

# Task 2b - Gamma curves normalized
x = np.arange(0, 1.001, 0.001)
for g in gamma_values:
    plt.plot(x, x ** g, label=f'={g}')
plt.title('Normalized (0-1)')
plt.legend(fontsize=7)
plt.show()

# Task 3 - Gamma correction on image
for i, g in enumerate([0.4, 1.0, 2.5]):
    result = (img.astype('float32') / 255.0) ** g
    plt.subplot(1, 3, i+1)
    plt.imshow(result, cmap='gray')
    plt.title(f'gamma={g}')
    plt.axis('off')
plt.show()

# Task 4 - Highlight range [A, B]
A, B = 100, 200
output = img.copy()
output[(img >= A) & (img <= B)] = 255
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(output, cmap='gray'); plt.title(f'[{A},{B}] -> 255'); plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = '/content/drive/MyDrive/opencv/download (1).jpeg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

for bit in range(1, 9):
    bit_plane = ((img >> (bit - 1)) & 1) * 255
    plt.subplot(2, 4, bit)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'bit plane{bit}')
    plt.axis('off')

plt.show()
```
