```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/opencv/fruit_basket.jpg')
print(image.shape)
```

```python
import numpy as np
translation_matrix = np.float32([[1, 0, 100], [0,
1, 50]])
```

```python
translated_image = cv2.warpAffine(image,
translation_matrix, (image.shape[1],
image.shape[0]))
print(translated_image.shape)
```

```python
plt.imshow(cv2.cvtColor(translated_image,
cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')
plt.show()
```

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
```

```python
rotation_matrix = cv2.getRotationMatrix2D(center,
45, 1.0)
```

```python
rotated_image = cv2.warpAffine(image,
rotation_matrix, (w, h))
```

```python
plt.imshow(cv2.cvtColor(rotated_image,
cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (45 degrees)')
plt.axis('off')
plt.show()
```

```python
Factor = 2
```

```python
scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0,
Factor)
```

```python
scaled_image = cv2.warpAffine(image,
scaling_matrix, (w*Factor, h*Factor))
```

```python
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (2x)')
plt.axis('off')
plt.show()
print(scaled_image.shape)
```

```python
m = 0.67
shearing_matrix = np.float32([[1, m, 0], [0, 1,
0]])
```

```python
sheared_image = cv2.warpAffine(image,
shearing_matrix, (w, h))
```

```python
plt.imshow(cv2.cvtColor(sheared_image,
cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m)
+')')
plt.axis('off')
plt.show()
```
