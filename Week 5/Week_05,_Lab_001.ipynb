```python
import cv2
import matplotlib.pyplot as plt

# Read a grayscale image
image = cv2.imread('/content/drive/MyDrive/download (2).jpeg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Create figure
plt.figure(figsize=(10, 5))

# ---- Original Image ----
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Histogram of Original Image
# ravel() flattens the 2D image matrix into a 1D array
# so we can compute its histogram
plt.subplot(2, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Histogram (Original)')

# ---- Equalized Image ----
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Histogram of Equalized Image
plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Histogram (Equalized)')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read color image
color_image = cv2.imread('/content/drive/MyDrive/opencv/images (1).jpeg')

# Split into B, G, R channels
b, g, r = cv2.split(color_image)

# Apply histogram equalization to each channel
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

# Merge channels back
equalized_color_image = cv2.merge([equalized_b, equalized_g, equalized_r])

# Display images
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')

# Equalized Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_color_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (Per Channel)')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt

# Read two grayscale images
image_source = cv2.imread('/content/drive/MyDrive/opencv/download (1).jpeg', cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('/content/drive/MyDrive/opencv/images (2).jpeg', cv2.IMREAD_GRAYSCALE)

# Perform histogram matching
matched_image = exposure.match_histograms(image_source, image_target)

# Display results
plt.figure(figsize=(15,5))

# Source Image
plt.subplot(3,3,1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')

plt.subplot(3,3,2)
plt.hist(image_source.ravel(), 256, [0,256])
plt.title('Histogram (Source)')

# Target Image
plt.subplot(3,3,4)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
plt.axis('off')

plt.subplot(3,3,5)
plt.hist(image_target.ravel(), 256, [0,256])
plt.title('Histogram (Target)')

# Matched Image
plt.subplot(3,3,7)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')

plt.subplot(3,3,8)
plt.hist(matched_image.ravel(), 256, [0,256])
plt.title('Histogram (Matched)')

plt.show()
```

```python
import cv2
from skimage import exposure
import matplotlib.pyplot as plt

# Read images in grayscale
source = cv2.imread('/content/drive/MyDrive/download (2).jpeg', cv2.IMREAD_GRAYSCALE)
target = cv2.imread('/content/drive/MyDrive/opencv/images (2).jpeg', cv2.IMREAD_GRAYSCALE)

# Step 1: Histogram Equalization
equalized_source = cv2.equalizeHist(source)

# Step 2: Histogram Matching on Equalized Image
matched_image = exposure.match_histograms(equalized_source, target)

# Display results
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(source, cmap='gray')
plt.title("Original Source")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(equalized_source, cmap='gray')
plt.title("After Equalization")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched_image, cmap='gray')
plt.title("After Equalization + Matching")
plt.axis('off')

plt.show()
```
